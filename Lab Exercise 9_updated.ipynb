```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score
```

```python
score_card = pd.DataFrame(columns=['Model No','Features','No of Features','Missing Featurs','Accuracy Score','Precision Score','Recall','F1','roc_auc'])
scalar = StandardScaler()
kmeans = KMeans(2)
for i in range(0,10):
    if i == 0:
        df_sample = df.copy()
    else:
        df_sample = df.sample(n=random.randint(8,15),axis='columns',replace=False)
    # display(df_sample.head())
    df_sample_cols = list(df_sample.columns)
    X = df.copy()
    y = df.index
    X = scalar.fit_transform(X)
    kmeans.fit(X)
    acc__scr = accuracy_score(y, kmeans.labels_)
    pre_scr = precision_score(y, kmeans.labels_)
    rec_scr = recall_score(y, kmeans.labels_)
    f1_scr = f1_score(y, kmeans.labels_)
    roc_auc = roc_auc_score(y, kmeans.labels_)
    score_card.loc[ len(score_card)] = ['Model_'+str(i+1),df_sample_cols,len(df_sample_cols),set(df)^set(df_sample_cols),acc__scr,pre_scr,rec_scr,f1_scr,roc_auc]

display(score_card)
```

inference:
Even after removing 9 features the level of accuracy remained consistent and no better results was obtained. Even after multiple iterations with different combinations the accuracy was around 0.5.
